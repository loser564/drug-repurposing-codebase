```python
import pandas as pd
import numpy as np
import yaml
```

```python
def load_config(yaml_path="P2-config.yaml"):
    with open(yaml_path, "r", encoding="utf-8") as f:
        return yaml.safe_load(f)
    
config = load_config()
```

```python
## get pdb and receptor info for diabetes drugs
all_drugs_df = pd.read_csv(config['all_drugs'])
for pair_cfg in config.get("disease_pairs", []):
    pair_name = pair_cfg.get("name", "unnamed_pair")
    # create emoty pandas df drug_info
    drug_info = {}

    # read chosen drugs
    df = pd.read_csv(pair_cfg["initial_pmid_list"])
    drugs = df['Drug'].unique().tolist()

    # remove epidermal growth factor drug
    if "epidermal growth factor" in drugs:
        drugs.remove("epidermal growth factor")

    # for each drug, get pdb and receptor info from all_drugs_df
    for drug in drugs:
        # make names smaller case for matching
        drug_lower = drug.lower()
        drug_row = all_drugs_df[all_drugs_df['DRUGNAME'].str.lower() == drug_lower]
        if not drug_row.empty:
            drug_name = drug_row['DRUGNAME'].values[0]
            pdb_id = drug_row['PDB_ID'].values[0]
            receptor = drug_row['TARGNAME'].values[0]
            drug_info[drug] = {'Drug': drug_name, 'Receptor': receptor, 'PDB_ID': pdb_id}
    # convert drug_info to pandas df
    drug_info_df = pd.DataFrame.from_dict(drug_info, orient='index')

    # save drug_info_df to csv
    drug_info_df.to_csv(pair_cfg["final_config"], index=False)







```
